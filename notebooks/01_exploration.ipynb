```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('../data/WESAD/S2/S2.pkl', 'rb') as f:
    data = pickle.load(f, encoding='latin1')
```

```python
print(type(data))
print(data.keys())
```

```python
print(data['signal'].keys())
print(data['signal']['chest'].keys())
print(data['signal']['wrist'].keys())
```

```python
print(data['label'].shape)
print(np.unique(data['label']))
```

```python
for signal_name, signal_data in data['signal']['chest'].items():
    print(f"Chest - {signal_name}: {signal_data.shape}")
```

```python
unique, counts = np.unique(data['label'], return_counts=True)

for label, count in zip(unique, counts):
    print(f"Label {label}: {count:,} samples")
```

```python
label_names = {
    0: 'Transient/Undefined',
    1: 'Baseline',
    2: 'Stress',
    3: 'Amusement',
    4: 'Meditation',
    5: 'Ignore',
    6: 'Ignore',
    7: 'Ignore'
}

for label, count in zip(unique, counts):
    name = label_names.get(label, 'Unknown')
    percentage = (count / len(data['label'])) * 100
    print(f"Label {label} ({name}): {count:,} samples ({percentage:.1f}%)")
```
